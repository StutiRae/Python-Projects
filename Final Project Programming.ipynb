## Cleaning the data

```python
#searching for all the nulls in the dataset
print("Missing values per column:\n", df.isnull().sum())
```

```python
#deleting the second column since it is empty
df = df.drop(df.columns[1],axis=1)
```

```python
#checking the dataset to ensure it is deleted
df
```

```python
#counting the nulls in the dataset
null_counts = df.isnull().sum()
#printing the columns which has nulls with the number of nulls in them
columns_with_nulls = null_counts[null_counts > 0]
print("Columns with null values:")
print(columns_with_nulls)
```

```python
#deleting the OWN_CAR_AGE column since it is mostly null and also irrelevant for our purpose
df = df.drop(columns = ['OWN_CAR_AGE'])
```

```python
#deleting the NAME_TYPE_SUITE column since it is irrelevant for our purpose
df = df.drop(columns = ['NAME_TYPE_SUITE']) 
```

```python
#deleting the OCCUPATION_TYPE column since it is mostly null and will not impact the meaning of the data
df = df.drop(columns = ['OCCUPATION_TYPE']) 
```

```python
df.info()
```

```python
#to handle the nulls in the AMT_ANNUITY column
#used the median to replace the null amount since the mean is sensitive to outliers
med_annuity=df['AMT_ANNUITY'].median()
```

```python
df['AMT_ANNUITY'].fillna(med_annuity,inplace=True)
```

```python
#now that the nulls for AMT_ANNUITY have been dealt with,
#doing the same with AMT_GOODS_PRICE and use the median to replace nulls
med_price = df['AMT_GOODS_PRICE'].median()
```

```python
df['AMT_GOODS_PRICE'].fillna(med_price,inplace=True)
```

```python
df.info()
```

```python
df = df.dropna(subset=['CNT_FAM_MEMBERS'])
```

```python
df.info()
```

```python
#checking for duplicates in SK_ID_CURR to ensure all values are unique
df['SK_ID_CURR'].duplicated().sum()
```

```python
#after having a look at the data, I noticed some rows have a "XNA" value which seems like a disguised null
#replacing the "XNA" with a null

df.replace("XNA", np.nan, inplace=True)
```

```python
#counting the nulls again in the dataset

null_counts = df.isnull().sum()

#printing the columns which has nulls with the number of nulls in them

columns_with_nulls = null_counts[null_counts > 0]
print("Columns with null values:")
print(columns_with_nulls)
```

```python
#deleting these rows from the dataset
df = df.dropna(subset=['CODE_GENDER'])
df = df.dropna(subset=['ORGANIZATION_TYPE'])
```
