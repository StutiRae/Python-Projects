```python
import pandas as pd
import plotly.graph_objs as go


delay_components = [
    'Delay Carrier (Minutes)',
    'Delay Weather (Minutes)',
    'Delay National Aviation System (Minutes)',
    'Delay Security (Minutes)',
    'Delay Late Aircraft Arrival (Minutes)'
]

fig = go.Figure()

for component in delay_components:
    fig.add_trace(go.Bar(
        x=df2['Date (MM/DD/YYYY)'],
        y=df2[component],
        name=component.split(' (')[0],  # Extracting component name without "(Minutes)"
        hoverinfo='y',
        marker=dict(line=dict(color='rgba(0,0,0,0.5)', width=0)),
        opacity=0.8,
    ))

fig.update_layout(
    barmode='stack',
    title='Stacked Bar Chart of Delay Components',
    xaxis_title='Date',
    yaxis_title='Delay (Minutes)',
    hovermode='x'
)

fig.show()
```

```python
import plotly.express as px
import pandas as pd

fig = px.scatter(df2, x='Scheduled Elapsed Time (Minutes)', y='Actual Elapsed Time (Minutes)',
                 color='Carrier Code', hover_name='Origin Airport',
                 title='Scheduled vs Actual Elapsed Time by Airline',
                 labels={'Scheduled Elapsed Time (Minutes)': 'Scheduled Time (min)',
                         'Actual Elapsed Time (Minutes)': 'Actual Time (min)',
                         'Airline': 'Airline'})

fig.update_xaxes(showgrid=True, gridwidth=0.5, gridcolor='lightgray')
fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor='lightgray')

fig.update_layout(
    hoverlabel=dict(bgcolor="white", font_size=12, font_family="Arial"),
    title_x=0.5,
    xaxis_title='Scheduled Time (min)',
    yaxis_title='Actual Time (min)',
    font=dict(family="Arial", size=12)
)
fig.show()

```

```python
import plotly.express as px
import pandas as pd

avg_elapsed_time = df2.groupby(['Carrier Code', 'Origin Airport'])['Actual Elapsed Time (Minutes)'].mean().reset_index()

fig = px.imshow(avg_elapsed_time.pivot(index='Origin Airport', columns='Carrier Code', values='Actual Elapsed Time (Minutes)'),
                labels=dict(x="Carrier Code", y="Origin Airport", color="Avg Elapsed Time (Minutes)"),
                x=avg_elapsed_time['Carrier Code'].unique(),
                y=avg_elapsed_time['Origin Airport'].unique(),
                color_continuous_scale='Viridis',
                title='Average Actual Elapsed Time by Airline and Origin Airport')

fig.update_layout(
    xaxis_title='Carrier Code',
    yaxis_title='Origin Airport',
    font=dict(family="Arial", size=12),
    coloraxis_colorbar=dict(title='Avg Elapsed Time (Minutes)'),
    width=1000,
    height=1000
)
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df1['Date (MM/DD/YYYY)'] = pd.to_datetime(df1['Date (MM/DD/YYYY)'])
df2['Date (MM/DD/YYYY)'] = pd.to_datetime(df2['Date (MM/DD/YYYY)'])

merged_df = pd.merge(df1, df2, on=['Carrier Code', 'Date (MM/DD/YYYY)', 'Flight Number', 'Tail Number'])
carriers = merged_df['Carrier Code'].unique()

plt.figure(figsize=(12, 6))

for carrier in carriers:
    carrier_data = merged_df[merged_df['Carrier Code'] == carrier]
    plt.scatter(carrier_data['Departure delay (Minutes)'], carrier_data['Arrival Delay (Minutes)'], label=carrier)

plt.xlabel('Departure Delay (Minutes)')
plt.ylabel('Arrival Delay (Minutes)')
plt.title('Departure Delay vs. Arrival Delay by Carrier')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df1['Date (MM/DD/YYYY)'] = pd.to_datetime(df1['Date (MM/DD/YYYY)'])
df2['Date (MM/DD/YYYY)'] = pd.to_datetime(df2['Date (MM/DD/YYYY)'])

total_delay_departures = df1.groupby('Date (MM/DD/YYYY)')['Departure delay (Minutes)'].sum()
total_delay_arrivals = df2.groupby('Date (MM/DD/YYYY)')['Arrival Delay (Minutes)'].sum()

plt.figure(figsize=(12, 6))
plt.plot(total_delay_departures.index, total_delay_departures.values, label='Departure Delays')
plt.plot(total_delay_arrivals.index, total_delay_arrivals.values, label='Arrival Delays')
plt.xlabel('Date')
plt.ylabel('Total Delay Time (Minutes)')
plt.title('Departure and Arrival Delays Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter data for carrier B6
b6_data = df1[df1['Carrier Code'] == 'B6']

# Group data by destination airport and calculate total delay count for each destination
delay_counts = b6_data['Destination Airport'].value_counts().head(20)

# Create a horizontal bar chart
plt.figure(figsize=(10, 8))
delay_counts.sort_values().plot(kind='barh', color='skyblue')
plt.xlabel('Number of Delays')
plt.ylabel('Destination Airport')
plt.title('Top 20 Destination Airports with the Most Delays for Carrier B6 (JetBlue Airways)')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Extract day of the week
df1['Day of Week'] = pd.to_datetime(df1['Date (MM/DD/YYYY)']).dt.day_name()

# Filter data for B6 flights to LAX
b6_lax_data = df1[(df1['Carrier Code'] == 'B6') & (df1['Destination Airport'] == 'LAX')]

# Group data by day of the week and calculate average delay
avg_delay_by_day = b6_lax_data.groupby('Day of Week')['Departure delay (Minutes)'].mean()

# Create a line plot
plt.figure(figsize=(10, 6))
avg_delay_by_day.sort_index().plot(kind='line', marker='o', color='skyblue')  # Sort index to ensure days are in the correct order
plt.xlabel('Day of the Week')
plt.ylabel('Average Departure Delay (Minutes)')
plt.title('Average Departure Delay for JetBlue Airways (B6) Flights to LAX by Day of the Week')
plt.xticks(rotation=0)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import plotly.graph_objs as go
import plotly.figure_factory as ff


# Select relevant delay columns
delay_columns = ['Delay Carrier (Minutes)', 'Delay Weather (Minutes)',
                 'Delay National Aviation System (Minutes)', 'Delay Security (Minutes)',
                 'Delay Late Aircraft Arrival (Minutes)',"Departure delay (Minutes)"]

# Calculate correlation matrix
correlation_matrix = df1[delay_columns].corr()

# Create heatmap of delay correlations
fig = ff.create_annotated_heatmap(
    z=correlation_matrix.values,
    x=delay_columns,
    y=delay_columns,
    colorscale='Viridis',
    annotation_text=correlation_matrix.round(2).values,
    showscale=True
)

fig.update_layout(
    title='Delay Correlations',
    xaxis_title='Delay Types',
    yaxis_title='Delay Types',
    xaxis=dict(side='bottom')
)

fig.show()
```
